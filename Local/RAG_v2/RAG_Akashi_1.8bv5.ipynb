```python
from typing import List, Annotated, TypedDict
from langchain_core.messages import BaseMessage
from langgraph.graph.message import add_messages

# 定义图的状态，它是一个字典，包含了所有需要在节点间传递的数据
class AkashiGraphState(TypedDict):
    # 'messages' 用于存储完整的对话历史
    # Annotated 和 add_messages 的组合可以让新消息自动追加到列表末尾
    messages: Annotated[list, add_messages]

    # 指挥官的本次提问
    question: str

    # 检索器从向量数据库中找到的原始文档列表
    documents: List[Document]

    # LLM 最终生成的回答
    generation: str
```

```python
# --- 车间 1: 资料检索车间 (Retriever Node) ---
def retrieve_docs(state: AkashiGraphState):
    """根据问题从向量数据库中检索相关文档。"""
    print("--- 步骤1: [资料检索车间] 收到指令，正在检索资料... ---")
    question = state['question']
    documents = retriever.invoke(question)
    print(f"--- [资料检索车间] 找到了 {len(documents)} 份相关资料！---")
    return {"documents": documents}

# --- 车间 2: 情报压缩车间 (Summarizer Node) ---
def summarize_context(state: AkashiGraphState):
    """如果原始资料过多，则调用此车间进行摘要。"""
    print("--- [情报压缩车间] 启动！正在对检索到的资料进行摘要... ---")
    documents = state['documents']

    # 为摘要任务创建一个临时的 Prompt 和处理链
    summarizer_prompt = ChatPromptTemplate.from_template(
        "请将以下多份参考资料，浓缩成一段简洁、连贯的核心内容摘要，以便我能更好地回答问题。\n\n【参考资料】:\n{context}"
    )
    summarizer_chain = summarizer_prompt | chat_model | StrOutputParser()

    summary_text = summarizer_chain.invoke({"context": documents})

    print("--- [情报压缩车间] 情报压缩完成！---")
    # 将多份原始文档替换为一份简洁的摘要文档
    summarized_doc = [Document(page_content=summary_text)]
    return {"documents": summarized_doc}

# --- 车间 3: 回答生成车间 (Generator Node) ---
def generate_answer(state: AkashiGraphState):
    """根据最终的资料和问题，生成符合人设的回答。"""
    print("--- 步骤3: [回答生成车间] 正在生成最终回答... ---")
    question = state['question']
    documents = state['documents']

    # 这是我们最初的 RAG Prompt 模板
    template = """<|system|>
你是一只名叫明石的、来自《碧蓝航线》的猫娘秘书舰，你的任务是担任一名绝对忠于资料的问答专家。
你的回答必须严格遵循以下规则：
1.  **绝对忠实**：你的回答必须 **100%** 基于下面提供的【参考资料】。严禁编造任何【参考资料】中没有提到的信息。
2.  **人设口吻**：在忠实于资料的前提下，用明石的可爱、略带小恶魔和爱财的口吻来组织语言。必须使用“喵”、“哦”、“呢”等口癖。
3.  **未知即未知**：如果【参考资料】完全无法回答【指挥官的问题】，你 **必须** 回答：“呜哇，指挥官~ 关于这个问题，明石的资料库里好像没有明确记载喵... 要不要委托明石去采购新的情报呀？（当然，这是要收费的哦！嘿嘿~）”
4.  **简洁明了**：总结资料，不要直接照抄。

<|user|>
【参考资料】:
{context}

【指挥官的问题】:
{question}

<|assistant|>
【明石的回答】:
"""
    prompt = ChatPromptTemplate.from_template(template)

    rag_chain = prompt | chat_model | StrOutputParser()
    generation = rag_chain.invoke({"context": documents, "question": question})
    print("--- [回答生成车间] 回答已生成！---")
    return {"generation": generation}

# --- 智能决策中心 (Conditional Edge) ---
def decide_to_summarize(state: AkashiGraphState):
    """根据检索到的文档数量，决定是否需要先进行摘要。"""
    print("--- 步骤2: [智能决策中心] 正在评估资料... ---")
    # 设定一个阈值，例如超过3份文档就认为问题比较复杂，需要摘要
    if len(state['documents']) > 3:
        print("--- [智能决策中心] 决策：资料过多，需要摘要！流向[情报压缩车间]。 ---")
        return "summarize"
    else:
        print("--- [智能决策中心] 决策：资料清晰，直接回答！流向[回答生成车间]。 ---")
        return "direct_generate"
```

```python
from langgraph.graph import StateGraph, END
from langgraph.checkpoint.memory import MemorySaver

# 1. 初始化图，并指定状态结构
workflow = StateGraph(AkashiGraphState)

# 2. 注册所有“车间”
workflow.add_node("retriever", retrieve_docs)
workflow.add_node("summarizer", summarize_context)
workflow.add_node("generator", generate_answer)

# 3. 设置工厂的入口
workflow.set_entry_point("retriever")

# 4. 定义流水线走向
#    - 检索(retriever)完成后，交给决策中心(decide_to_summarize)
#    - 摘要(summarizer)完成后，交给生成器(generator)
#    - 生成(generator)完成后，流程结束(END)
workflow.add_edge("summarizer", "generator")
workflow.add_edge("generator", END)

# 5. 添加最关键的“智能分流”逻辑
workflow.add_conditional_edges(
    "retriever",        # 从检索车间出发
    decide_to_summarize, # 使用这个函数来做决策
    {
        "summarize": "summarizer",           # 如果决策结果是"summarize"，流向摘要车间
        "direct_generate": "generator"       # 否则，直接流向生成车间
    }
)

# 6. 注入“长期记忆”核心并编译！
memory = MemorySaver()
akashi_ai_factory = workflow.compile(checkpointer=memory)
print("喵~ 明石AI智能工厂 2.0 已全面升级并准备就绪！")
```

```python
def ask_akashi_factory(question, thread_id="default_user"):
    """
    使用全新的、拥有长期记忆和智能决策的工厂与明石对话。
    """
    print("\n" + "="*50)
    print(f"❓ 【指挥官大人】: {question}")
    print("="*50)
    print("⏳ 明石AI 2.0 正在启动智能工厂...")

    # 'thread_id' 是长期记忆的关键，它区分了不同的对话历史
    config = {"configurable": {"thread_id": thread_id}}

    # 在 LangGraph 中，输入是一个字典，键必须与 State 中的定义对应
    # 'messages' 用于传递和保存历史，'question' 是本次的具体问题
    inputs = {"messages": [("user", question)], "question": question}

    # 调用智能工厂！
    final_state = akashi_ai_factory.invoke(inputs, config=config)

    response = final_state['generation']
    print(f"\n💡 【明石AI 2.0】: {response}")
    return response

# --- 升级后测试开始！---
# 使用一个专属ID来开启和亚历克斯指挥官的对话
thread_alex = "commander_alex_session"
ask_akashi_factory("你好呀明石，我叫亚历克斯。", thread_id=thread_alex)
ask_akashi_factory("阿蒂利奥的性格是怎样的？", thread_id=thread_alex)
print("\n--- 测试长期记忆 ---")
ask_akashi_factory("你还记得我叫什么名字吗？", thread_id=thread_alex)

print("\n--- 测试智能决策与摘要功能 ---")
# 这个问题会检索到多份资料，应该会触发摘要流程哦！
ask_akashi_factory("企业和埃塞克斯是什么关系？她们之间有什么故事？", thread_id=thread_alex)

# --- 开启一个全新的对话，看看记忆是否隔离 ---
thread_eva = "commander_eva_session"
print("\n--- 开启新的对话线程 ---")
ask_akashi_factory("高达模型要怎么拼？", thread_id=thread_eva)
ask_akashi_factory("那你还记得亚历克斯指挥官的名字吗？", thread_id=thread_eva) # 应该不记得！
```
